```python
# Scrape URL: 
base_url = 'https://www.immobiliare.it/en/mercato-immobiliare/'

# References
#https://hackersandslackers.com/scraping-urls-with-beautifulsoup
```

```python
import requests
from bs4 import BeautifulSoup, Tag
import pandas as pd

headers = {
    'Access-Control-Allow-Origin': '*',
    'Access-Control-Allow-Methods': 'GET',
    'Access-Control-Allow-Headers': 'Content-Type',
    'Access-Control-Max-Age': '3600',
    'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/52.0'
    }
```

```python
req = requests.get(base_url, headers)
soup = BeautifulSoup(req.content, 'html.parser')
#print(soup.prettify())
```

```python
#look for 'table' class: nd-table nd-table--borderBottom
region_table = soup.find_all("table", class_="nd-table nd-table--borderBottom")
```

```python
#look for the 'tr' class: nd-table__row
tr_tags = soup.find_all("tr", class_="nd-table__row")
#tr_tags
```

```python
a = 0
b = 0
c = 0

for tag in tr_tags:
    print(f'A{a}: type(tag): {type(tag)}')
    print(f'A{a}: tag: {tag}')
    for z in tag.children:
        
        if type(z) == Tag:
            
            print(f'C{c}::{z.get("data-text")}::{z.text}::') 
            
            w = z.find('a')
            if (w):
                link = w.get('href')
                text = w.text
                print(f'text/link: {text}/{link}')
        c+=1
        b += 1
        
    print('-'*10)
    a += 1

    
```

```python
results = []

tr_tags = soup.find_all("tr", class_="nd-table__row")


for tr_tag in tr_tags:
    loop = 0
    
    region = ''
    url = ''
    sale = 0
    rent = 0


    for child in tr_tag.children:
        if type(child) == Tag:
            
            if child.get("data-text") == "(€/m²)":
                link = child.find('a')
                if (link):
                    url = link.get('href')
                    region = link.text
#                    print(f'{loop}::text/link: {link_text}/{link_url}')

            elif child.get("data-text") == "Sale":
                sale = child.text
            elif child.get("data-text") == "Rent":
                rent = child.text
                
            
            
#            print(f'{loop}>{child.get("data-text")}<{child.text}::') 
            
                
        loop += 1
    print(f'{region}-{sale}-{rent}-{url}')
    results.append([region, sale, rent, url])
    print('-'*10)
    
results
results_df = pd.DataFrame(data = results, columns=['region', 'sale', 'rent', 'url'])
results_df
```
