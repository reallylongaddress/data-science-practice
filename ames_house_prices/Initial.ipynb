```python
import pandas as pd

df_train = pd.read_csv('data/train.csv')
df_train.head()
```

```python
X_train = df_train.copy().drop(columns='SalePrice')
y_train = df_train.copy()['SalePrice']
print(f'X_train.shape: {X_train.shape}')
```

```python
X_train.head()
```

```python
X_train.isnull().sum().sort_values(ascending=False).head(20)
```

# Baseline model #

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_validate
```

```python
base_linear_regression_model = LinearRegression()
base_cv_results = cross_validate(base_linear_regression_model, X_train, y_train, cv=5, scoring=('r2', 'neg_mean_squared_error'))

base_r2 = base_cv_results['test_r2'].mean()
base_neg_mean_squared_error = base_cv_results['test_neg_mean_squared_error'].mean()

#print(f'cv_results: {cv_results}')
print(f'base_r2 mean: {base_r2}')
print(f'base_neg_mean_squared_error mean: {base_neg_mean_squared_error}')
```
